```python
#preliminary visualization 1
data['auto_make'].value_counts().plot(kind='bar',
color='b')
plt.xlabel('Car Companies')
plt.ylabel('count')
plt.title('number of cars by each company')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#preliminary visualization 2
plt.hist(data['total_claim_amount'])
plt.xlabel('total amount claimed')
plt.ylabel('count of total amout claimed ')
plt.title('number of amout claimed  ')
plt.tight_layout()
plt.show()
```
